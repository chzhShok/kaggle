```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
data = pd.read_csv('train.csv', index_col='PassengerId')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
sns.heatmap(data.corr(), annot=True, square=True, fmt='.2f')
```

```python
data = data.drop(columns=['Name', 'Ticket', 'Cabin'])
```

```python
data['Sex'] = data['Sex'].replace(['male', 'female'], [0, 1])
data
```

```python
sns.histplot(x='Fare', y='Survived', data=data)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].value_counts().idxmax())
```

```python
data['Embarked']
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)

cat_data = data[['Embarked']]
num_data = data.drop(columns=['Embarked'])
```

```python
one_hot_data = enc.fit_transform(cat_data)

one_hot_data = pd.DataFrame(one_hot_data, index=cat_data.index)

data = pd.concat([num_data, one_hot_data], axis=1)

data
```

```python
y = data[['Survived']]
X = data.drop(columns=['Survived'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8)

X_train.columns = X_train.columns.astype(str)
X_val.columns = X_val.columns.astype(str)
```

# **Обучение модели**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

# Логистическая регрессия

```python
from sklearn.preprocessing import normalize

normalized_train = normalize(X_train)
normalized_val = normalize(X_val)

norm_train_data = pd.DataFrame(normalized_train)
norm_val_data = pd.DataFrame(normalized_val)
```

```python
from numpy.core.function_base import logspace
params_lg = {'penalty':('l1', 'l2', None), 'C': np.logspace(-5, 5, 11)}

lr = LogisticRegression()
grid_log = GridSearchCV(lr, params_lg, cv=5)

grid_log.fit(norm_train_data, y_train)
grid_log.best_params_
```

```python
# {'C': 1000.0, 'penalty': 'l2'}

lr = LogisticRegression(C=1000, penalty='l2')

lr.fit(norm_train_data, y_train)

y_pred_log = lr.predict(norm_val_data)
```

```python
accuracy_score(y_val, y_pred_log)
```

```python
#0.7988826815642458
```

# K-соседей

```python
params_kneighbors = {'n_neighbors':np.arange(3, 15)}

kn = KNeighborsClassifier()
grid_kneighbors = GridSearchCV(kn, params_kneighbors, cv=5)

grid_kneighbors.fit(X_train, y_train)
grid_kneighbors.best_params_
```

```python
# {'n_neighbors': 8}
kn = KNeighborsClassifier(n_neighbors=8)

kn.fit(X_train, y_train)

y_pred_kneighbors = kn.predict(X_val)
```

```python
accuracy_score(y_val, y_pred_kneighbors)
```

```python
# 0.6759776536312849
```

# Решающее дерево

```python
params_dtc = {'criterion':('gini', 'entropy', 'log_loss'),
              'max_depth': np.arange(5, 15),
              'min_samples_leaf': np.arange(1, 10),
              'min_samples_split':np.arange(2, 11)}

dtc = DecisionTreeClassifier()
grid_dtc = GridSearchCV(dtc, params_dtc, cv=5)

grid_dtc.fit(X_train, y_train)
```

```python
grid_dtc.best_params_
```

```python
dtc = DecisionTreeClassifier(criterion='gini',
                             max_depth=10,
                             min_samples_leaf=6,
                             min_samples_split=2)

dtc.fit(X_train, y_train)

y_pred_dtc = dtc.predict(X_val)
```

```python
accuracy_score(y_val, y_pred_dtc)
```

```python
# 0.7932960893854749
```

# Случайный лес

```python
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier()
depths = np.arange(3, 20)
rf_metrics = []

for depth in depths:
    scores = cross_val_score(rfc, X_train, y_train, cv=5)

    rf_metrics.append(np.mean(scores))
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(3, 20), rf_metrics)
plt.xlabel('depth')
plt.ylabel('roc-auc-score')
plt.grid()
```

Выбор лучших парамтеров Случайного леса

```python
params_rfc = {'n_estimators': range (10, 51, 10),
              'max_depth': range (1,13, 2),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2),
              'criterion':('gini', 'entropy', 'log_loss')}

rfc = RandomForestClassifier()
grid_rfc = GridSearchCV(rfc, params_rfc, cv=5)

grid_rfc.fit(X_train, y_train)
```

```python
grid_rfc.best_params_
```

```python
# {'criterion': 'entropy',
 #'max_depth': 5,
 #'min_samples_leaf': 2,
 #'min_samples_split': 6,
 #'n_estimators': 20}

rfc = RandomForestClassifier(criterion = 'entropy',
                              max_depth=5,
                              min_samples_leaf=2,
                              min_samples_split=6,
                              n_estimators=20)

rfc.fit(X_train, y_train)

y_pred_rfc = rfc.predict(X_val)
```

```python
accuracy_score(y_val, y_pred_rfc)
```
